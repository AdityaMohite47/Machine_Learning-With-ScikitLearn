<h3> Exploring Support Vector Classifier </h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset is about the infected mouse been given doses to recover
# and after 2 weeks the mouse's codition is checked ( Traget variable )

data = pd.read_csv('./Datasets/mouse_viral_study.csv')
```

```python
data.sample(10)
```

```python
# Perfect seperated classes 
sns.scatterplot(x='Med_1_mL' , y='Med_2_mL' , data=data , hue='Virus Present')
```

-  Making model with Support Vector Classifier to visualize hyperplane

```python
from sklearn.svm import SVC
X = data.drop('Virus Present' , axis=1)
y = data['Virus Present']

model = SVC(kernel='linear' , C=1000)
model.fit(X , y)
```

```python
# This 3-rd party function helps us to visualize hyperplane made by the kernel 
# It also highlights the support vectors and margins/soft-margins...

from svm_margin_plot import plot_svc_boundary

plot_svc_boundary(model , X , y) 
```

- Notes on Hyper-Parameters of Support Vector Machines

    * epsilon : The amount of error allowed , misclassifications allowed in the margin

    * kernel : Computation machines that visualize the data in higher dimensions , where it's easy to apply hyperplane seperation.

    * C : Denotes the Regularization Strength.

    * Gamma ( only when kernel is rbf ) : Denotes how much influence a data point has while mapping the hyperplane. 

( in sklearn application is inverted i.e bigger the value less the limit and vise-versa)
